# Removing urls
## with the help of regular expressions we can remove urls

```python

 
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python

text1 = 'Hello I am jay https://www.google.com/'
text2 = 'Hi I am jay www.google.com'
text3 = 'Hola '
```

```python
remove_url(text1)
```

```python
remove_url(text2)
```

```python
remove_url(text3)
```

# REMOVING PUNCTUATIONS

```python

import string
string.punctuation
```

```python
exclude = string.punctuation
def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
str1 = 'Hi, How are you? Hope you are fine!'
remove_punc(str1)
```

# CHAT WORD TREATMENT

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible',
    'ATK':'At The Keyboard',
    'ATM':'At The Moment'
}
```

```python
def replace_chat(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
replace_chat('AFAIK he is a good man')
```

```python
replace_chat('ATM I''m broke')
```

# Removing stopwords

```python
from nltk.corpus import stopwords
```

```python
#these are the stopwords in english
stopwords.words('english')
```

```python
text = 'probably my all time favorate movie, is narsimham'
#to remove stopwords from a sentence,
def remove_stopwords(text):
    new_text=[]
    for w in text.split():
        if w in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(w)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
        
        
```

```python
remove_stopwords(text)
```

# Handling Emojis
 ## Removing with regex

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U0001F004-\U0001F0CF"  # Additional emoticons
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'walk walking walks walked'
stem_words(sample)
```

```python
text = """Inflectional changes are an important aspect of grammar in many languages. They convey grammatical information that helps clarify the relationships between words in a sentence and provide context for understanding the meaning of a sentence."""
stem_words(text)
#in stemming while we stem the root word it may not have a valid meaning
#to overcome this we use lemmatization
```

# Lemmatization

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp("eating eat ate eats adjustable rafting running varying meeting better")
for token in doc:
    print(token,"|",token.lemma_,"|",token.lemma)
```
