# PART OF SPEECH(POS)

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
nlp.pipe_names
```

```python
doc = nlp("Elon flew to mars yesterday. He carried biriyani masala with him")
for token in doc:
    print(token,"|",token.pos_,"|",spacy.explain(token.pos_))
```

```python
doc = nlp("Elon flew to mars yesterday. He carried biriyani masala with him")
for token in doc:
    print(token,"|",token.pos_,"|",token.lemma_)
```

```python
#tags in spacy
#we can know that each tokens are in past tense or present etc
# for that we use tags

doc = nlp("Elon flew to mars yesterday. He carried biriyani masala with him")
for token in doc:
    print(token,"|",token.pos_,"|",token.lemma_,"|",token.tag_,"|",spacy.explain(token.tag_))
```

## Exercise

```python
story = """THE whole sky had been overcast with rain-clouds from early morning; it was a still day, not hot, but heavy, as it is in grey dull weather when the clouds have been hanging over the country for a long while, when one expects rain and it does not come. Ivan Ivanovitch, the veterinary surgeon, and Burkin, the high-school teacher, were already tired from walking, and the fields seemed to them endless. Far ahead of them they could just see the windmills of the village of Mironositskoe; on the right stretched a row of hillocks which disappeared in the distance behind the village, and they both knew that this was the bank of the river, that there were meadows, green willows, homesteads there, and that if one stood on one of the hillocks one could see from it the same vast plain, telegraph-wires, and a train which in the distance looked like a crawling caterpillar, and that in clear weather one could even see the town. Now, in still weather, when all nature seemed mild and dreamy, Ivan Ivanovitch and Burkin were filled with love of that countryside, and both thought how great, how beautiful a land it was."""
```

```python
story = story.lower()
story
```

```python
# removing the spaces and punctiations and printing the filtered tokens
doc = nlp(story)
filtered_tokens = []
for token in doc:
    if token.pos_ not in ["SPACE","PUNCT","X"]:
            filtered_tokens.append(token)

for words in filtered_tokens:
    print(words)
```

```python
#taking the count of pos in this tokens
count = doc.count_by(spacy.attrs.POS)
count
```

```python
for k,v in count.items():
    print(doc.vocab[k].text,"|",v)
```

```python
doc4 = nlp("Bloomberg  is going to acquire Brototype for $45 billion")
for ent in doc4.ents:
    print(ent.text,"|",ent.label_,"|",spacy.explain(ent.label_))
    
#now bloomberg is person not org,so we need to customize
```

```python
# when we want to customize our spacy use span
type(doc4[0:1])
```

```python
from spacy.tokens import Span
s1 = Span(doc4,0,1,label="ORG")
doc4.set_ents([s1],default="unmodified")
```

```python
for ent in doc4.ents:
    print(ent.text,"|",ent.label_,"|",spacy.explain(ent.label_))
#now the bloomberg is an org   
```
